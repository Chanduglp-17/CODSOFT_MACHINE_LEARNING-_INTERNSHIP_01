```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset
data = pd.DataFrame({
    'V1': [1.2, 0.3, 1.5, 0.1, 0.7],
    'V2': [0.4, 0.2, 0.6, 0.3, 0.8],
    'Class': [0, 1, 0, 1, 0]
})

# Save to CSV
data.to_csv('credit_card_transactions.csv', index=False)

# Load the dataset
data = pd.read_csv('credit_card_transactions.csv')

# Calculate the number of transactions by type
transaction_counts = data['Class'].value_counts()

# Prepare labels and data for the pie chart
labels = ['Legitimate', 'Fraudulent']
sizes = [transaction_counts.get(0, 0), transaction_counts.get(1, 0)]
colors = ['#66c2a5', '#fc8d62']  # Green for legitimate, Red for fraudulent

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Credit Card Transactions')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('credit_card_transactions.csv')

# Calculate the number of transactions by type
transaction_counts = data['Class'].value_counts()

# Create a DataFrame for displaying
table_data = pd.DataFrame({
    'Transaction Type': ['Legitimate', 'Fraudulent'],
    'Count': [transaction_counts.get(0, 0), transaction_counts.get(1, 0)]
})

# Plot the table
fig, ax = plt.subplots(figsize=(6, 2))  # Set the size of the plot
ax.axis('off')  # Hide the axes

# Create the table
table = ax.table(cellText=table_data.values,
                 colLabels=table_data.columns,
                 cellLoc='center',
                 loc='center',
                 rowColours=['#f1f1f2', '#f1f1f2'],
                 colColours=['#40466e'] * len(table_data.columns),
                 cellColours=[['#ffffff'] * len(table_data.columns)] * len(table_data),
                 bbox=[0, 0, 1, 1])  # Position and size of the table

plt.title('Transaction Counts')
plt.show()
```
